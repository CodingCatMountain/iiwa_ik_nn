```python
### import library
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
print(tf.__version__)
```

```python
df = pd.read_csv('../bag/iiwa_samples_with_noise.csv')
df.head()
```

```python
print(df.info())
## print(df.describe())
```

```python
#### 提取 7个关节角度作为label 7个 pos作为数据量
import ast
Labels = [];Datas =[]
for row in df.itertuples():
    Link_pos_1 =  ast.literal_eval(row.link_pos_1)
    Link_pos_2 =  ast.literal_eval(row.link_pos_2)
    Link_pos_3 =  ast.literal_eval(row.link_pos_3)
    Link_pos_4 =  ast.literal_eval(row.link_pos_4)
    Link_pos_5 =  ast.literal_eval(row.link_pos_5)
    Link_pos_6 =  ast.literal_eval(row.link_pos_6)
    Link_pos_7 =  ast.literal_eval(row.link_pos_7)
    Pose = [Link_pos_1,Link_pos_2,Link_pos_3,Link_pos_4,Link_pos_5,Link_pos_6,Link_pos_7]
    #Pose=[Link_pos_7]
    j1 = row.iiwa_joint_1
    j2 = row.iiwa_joint_2
    j3 = row.iiwa_joint_3
    j4 = row.iiwa_joint_4
    j5 = row.iiwa_joint_5
    j6 = row.iiwa_joint_6
    Joint = [j1,j2,j3,j4,j5,j6]
    
    Datas.append(Pose)
    Labels.append(Joint)
```

```python
dict ={"Joint":Labels,"Pose":Datas}
df_out = pd.DataFrame(dict)
df_out.to_csv('./iiwa_dataset.csv')
```
